```python
#and more importantly: these package 
#"call" libraries written in C or Fortran (fast),
#are written by people who are really good in coding and found super smart tricks
# ====> they are MUCH faster than what we can write in few lines of code
l = np.array([1,2,3])
k = 5
n_test = 1000

def func_me():
    for _ in range(len(l)):
        l[_] *= k #note here that we use a more compact syntax
    return l
time_me = timeit(func_me, number=n_test)/n_test
print('My function took', time_me, 's')


def func_numpy():
    return k*l
time_numpy = timeit(func_numpy, number=n_test)/n_test
print('Numpy took', time_numpy, 's')
```

```python
#let's do a density profile
#we need the distance
r = (x_DM**2 + y_DM**2 + z_DM**2)**0.5
#the mass
m = ds.r[('DM','particle_mass')].to('Msun')
```

```python
#maybe look how r looks like
plt.figure()
plt.plot(r, linestyle='', marker=',')

plt.figure()
plt.plot(r, linestyle='', marker=',')
plt.yscale('log') #set the y-axis in log-scale
plt.grid(axis='y', which='both') #add grid
```

```python
#we will need to bin in r and measure what is the amount of mass
#in each bins, then we can divide by the volume of the bin to obtain
#a density profile

#bins
bin_r = np.logspace(1,5,100)
print('bin_r:',bin_r)
```

```python
np.histogram?
```

```python
#No need to recode the histogram, smart people have already done it!
mass_bins, bin_r = np.histogram(r, bins=bin_r, weights=m)
print('lenghts:',len(mass_bins), len(bin_r))
```

```python
#look the density profile
rho = mass_bins / np.diff(4/3*np.pi*bin_r**3)
bin_r = (bin_r[1:] + bin_r[:-1])/2

plt.loglog(bin_r, rho)
#note the faster way to plot in log log
#one could do plt.plot and then plt.xscale('log') and plt.yscale('log')
plt.xlabel('$r$ [pc]') #$ is used to put some latex
plt.ylabel('$\\rho$ [Msun/pc**3]')
```

```python
#Most halos can be fitted with a Navarro Frenk and White profile
# https://en.wikipedia.org/wiki/Navarro%E2%80%93Frenk%E2%80%93White_profile
#is it the case for this one? can we infer the two parameters (rho0 and rs)?
#interlude on numerical fitting
```

```python
scipy.optimize.curve_fit?
```

```python
#define the functional to fit
def NFW(r, rho0, rs):
    return rho0/(r/rs * (1+ r/rs)**2)

#filt on bad values of the density profile
filt = (bin_r >200) & (bin_r < 3e4) 
```

```python
plt.loglog(bin_r, rho, linewidth=5)
plt.loglog(bin_r[filt], rho[filt], linestyle='--', color='k')
plt.xlabel('$r$ [pc]')
plt.ylabel('$\\rho$ [Msun/pc**3]')
```

```python
plt.loglog(bin_r, rho, linewidth=5)
plt.loglog(bin_r[filt], rho[filt], linestyle='--', color='k', label='data')

#popt,pcov = scipy.optimize.curve_fit(NFW, bin_r[filt], rho[filt])
popt,pcov = scipy.optimize.curve_fit(NFW, bin_r[filt], rho[filt], maxfev=10000)
plt.loglog(bin_r, (NFW(bin_r, popt[0], popt[1])), label='fit lin')

plt.xlabel('$r$ [pc]')
plt.ylabel('$\\rho$ [Msun/pc**3]')
plt.legend()
```

```python
plt.loglog(bin_r, rho, linewidth=5)
plt.loglog(bin_r[filt], rho[filt], linestyle='--', color='k', label='data')

#popt,pcov = scipy.optimize.curve_fit(NFW, bin_r[filt], rho[filt])
popt,pcov = scipy.optimize.curve_fit(NFW, bin_r[filt], rho[filt], maxfev=10000)
plt.loglog(bin_r, (NFW(bin_r, popt[0], popt[1])), label='fit lin')

def log_NFW(r, rho0, rs):
    return np.log(rho0/(r/rs * (1+ r/rs)**2))
popt,pcov = scipy.optimize.curve_fit(log_NFW, bin_r[filt], np.log(rho)[filt])
plt.loglog(bin_r, np.exp(log_NFW(bin_r, popt[0], popt[1])), label='fit log')

plt.xlabel('$r$ [pc]')
plt.ylabel('$\\rho$ [Msun/pc**3]')
plt.legend()
```
